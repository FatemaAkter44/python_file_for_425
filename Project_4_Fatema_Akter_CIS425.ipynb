```python
keyChar2Darray = ["", "", "ABC", "DEF", "GHI", "JKL", "MNO", "PQRS", "TUV", "WXYZ"] 

#keyChar2Darray[0][0]=
#keyChar2Darray[0][1]=
#keyChar2Darray[1][0]=
#keyChar2Darray[1][1]=
#keyChar2Darray[1][2]=
#keyChar2Darray[2][0]=A
#keyChar2Darray[2][1]=B
#keyChar2Darray[5][2]=K

def printWordsUtil(number, curr, output, n): #
    if(curr == n): 
        print(output) 
        return

    for i in range(len(keyChar2Darray[number[curr]])): 
        output.append(      keyChar2Darray     [number[curr]]      [i]     ) 
        printWordsUtil(number, curr + 1, output, n) 
        output.pop() # repeate char delete
        if(number[curr] == 0 or number[curr] == 1): 
            return;  
             
def printWords(number, n): 
    printWordsUtil(number, 0, [], n) 
      
if __name__ == '__main__':
    number = [2,3,5,7,7,7,3] 
    n = len(number) 
    printWords(number, n); #n=2 number[0]=2 number[1]=3
```
