```python
principle = 1000
anual_interest = .05
for years in range (1,11):
    amount = principle * (1 + anual_interest) ** years
    #print(f'{"years":>2}{"amount":>20.2f}')
    print(f'{years:>2}{amount:>20.2f}')
```
